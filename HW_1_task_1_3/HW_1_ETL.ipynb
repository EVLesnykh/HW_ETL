```python
task3_df = pd.DataFrame({
    'Employee_ID': ['E001', 'E001', 'E002', 'E002', 'E003'],
    'Name': ['Alice', 'Alice', 'Bob', 'Bob', 'Alice'],
    'Job_сode': ['J01', 'J02', 'J02', 'J03', 'J01'],
    'Job': ['Chef', 'Waiter', 'Waiter', 'Bartender', 'Chef'],
    'City_code': [26, 26, 56, 56, 56],
    'Home_city': ['Moscow', 'Moscow', 'Perm', 'Perm', 'Perm']
})
task3_df
```

Данная таблица находится в 1NF

**Приведение таблицы к 2 NF**



```python
task3_employees = task3_df[['Employee_ID', 'Name', 'City_code', 'Home_city']].drop_duplicates().reset_index(drop=True)
task3_employees
```

```python
task3_job_roles = task3_df[['Employee_ID', 'Job_сode', 'Job']]
task3_job_roles
```
